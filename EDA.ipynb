```python
import spacy
import pandas as pd
import asyncio
```

```python
df = pd.read_csv('IMDB Dataset.csv')
df.head()
```

```python
nlp = spacy.load("en_core_web_sm")
doc = nlp("Apple is looking at buying U.K. startup for $1 billion")
for token in doc:
    print(token.text, token.pos_, token.dep_)
```

```python
for ent in doc.ents:
    print(ent.text, ent.label_)
```

```python
def get_nouns(lists, nouns):
    for i in lists:
        doc = nlp(i)
        for ent in doc.ents:
            # print(ent.text, ent.label_)
            if ent.label_ == 'PERSON':
                nouns.append([ent.text, 'Person'])
            elif ent.label_ == 'GPE':
                nouns.append([ent.text, 'City'])
    return nouns
```

```python
nouns = []
for i in range(0, len(df), 100):
    print(i)
    get_nouns(df['review'][i:i+100], nouns)
```

```python
df_nouns = pd.DataFrame(data=nouns, columns=['Word', 'Type'])
```

```python
j_dict = {}
types = ['City', 'Person']
for i in types:
    print(i)
    j_dict[i] = list(df_nouns[df_nouns['Type']==i]['Word'])
print(j_dict)
```

```python
import json
json_object = json.dumps(j_dict)
with open('nouns.json', 'w') as json_file:
    json_file.write(json_object)
```
